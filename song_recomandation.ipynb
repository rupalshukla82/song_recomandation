```python
import numpy as np
import pandas as pd
import Recommenders as Recommenders
```

```python
song_Df1 = pd.read_csv('triplets_file.csv')
song_Df1.head()
```

```python
song_Df2 = pd.read_csv('song_data.csv')
song_Df2.head()
```

```python
# merge both the data into the single datafrme
song_Df = pd.merge(song_Df1,song_Df2.drop_duplicates(['song_id']),on='song_id',how='left')
song_Df.head()
```

```python
# merging the song title and artisit name by using - 
song_Df['song'] = song_Df['title']+' - '+ song_Df['artist_name']
song_Df.head()
```

```python
song_Df = song_Df.head(10000)
```

```python
song_Df['song']
```

```python
# comulative of sum count
song_grouped = song_Df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
# grouped by sum in the persantage
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = ((song_grouped['listen_count']) / grouped_sum)*100
song_grouped.sort_values(['listen_count','song'],ascending=[0,1])
```

```python
# popularity recomandation engine 
#  music recomandation base on the popularity
pr = Recommenders.popularity_recommender_py() 
pr.create(song_Df,'user_id','song')
pr.recommend(song_Df['user_id'][5])
```

```python
# item simarity enigine 
# base on the user listening recomnade the song 
ir = Recommenders.item_similarity_recommender_py()
ir.create(song_Df,'user_id','song')
```

```python
user_items = ir.get_user_items(song_Df['user_id'][100])
```
